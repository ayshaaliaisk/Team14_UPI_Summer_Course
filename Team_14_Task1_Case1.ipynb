# Case Study in Python Programming - Task 1

## First Case Study : Miles PerGallon

- Drivers are concerned with the mileage obtained by their automobiles. One driver has kept track of several tankfuls of gasoline
by recording miles driven and gallons used for each tankful. Develop a sentinel-controlled-repetition script that prompts the
user to input the miles driven and gallons used for each tankful.
- The script should calculate and display the miles per gallon obtained for each tankful. After processing all input information, the
script should calculate and display the combined miles per gallon obtained for all tankfuls (that is, total miles driven divided by
total gallons used).
- Requirment Output of The Program:

image.png

```python
def main():
    # Initialize variables to store total miles and total gallons
    total_miles = 0
    total_gallons = 0

    while True:
        # Ask the user to enter the gallons used
        gallons = float(input("Enter the gallons used (-1 to end): "))
        # If the input is -1, exit the loop
        if gallons == -1:
            break

        # Ask the user to enter the miles driven
        miles = float(input("Enter the miles driven: "))
        # If the input is -1, exit the loop
        if miles == -1:
            break

        # Calculate the miles per gallon for the current tank
        miles_per_gallon = miles / gallons
        # Display the miles per gallon with 6 decimal
        print(f"The miles/gallon for this tank was {miles_per_gallon:.6f}")

        # Add the current miles and gallons to the totals
        total_miles += miles
        total_gallons += gallons

    # Loop ends, calculate and display the overall average miles per gallon
    if total_gallons != 0:
        combined_miles_per_gallon = total_miles / total_gallons
        print(f"The overall average miles/gallon was {combined_miles_per_gallon:.6f}")
    else:
        # Error handling if no data was entered, display this message
        print("No data to calculate combined miles per gallon.")

# Main Function
if __name__ == "__main__":
    main()
```
